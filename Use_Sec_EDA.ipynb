```python
def plot_hist(df,col,xmax):
  fig,ax=plt.subplots(figsize=(6,4))
  x=df[col]
  x.plot(kind='hist',density=True,bins=10)
  x.plot(kind='kde')

  ax.set_xlabel('Number of '+col)
  ax.set_ylabel('Frequency')
  ax.set_title(col+ ' Histogram')
  # plt.xticks(np.arange(0,xmax,10))
  plt.xlim([0,xmax])
  plt.show()
```

```python
#comparing all 4 together
def plot_hist_combined(df1,df2,df3,df4,col,xmax):
  fig,ax=plt.subplots(figsize=(6,4))
  x1=df1[col]
  x1.plot(kind='kde', label = 'USA')
  x2 = df2[col]
  x2.plot(kind='kde', label = 'India')
  x3= df3[col]
  x3.plot(kind='kde', label = 'Japan')
  x4= df4[col]
  x4.plot(kind='kde', label = 'Netherlands')
  ax.set_xlabel('Number of '+col)
  ax.set_ylabel('Frequency')
  ax.set_title(col+ ' Histogram')
  plt.legend(loc="upper right")
  plt.xlim([0,xmax])
  plt.show()
```

```python
import operator
def sensitive_eda(dfN):
  df = dfN[dfN['Sensitive'].notna()]
  sens_agg_list=[]
  for index, row in df.iterrows():
    semis=row['Sensitive']
    list_sens=semis.split(';')
    sens_agg_list.append(list_sens)
  dict_sens={}
  
  for sens_list in sens_agg_list:
    for i in sens_list:
      if i in dict_sens.keys():
        dict_sens[i]+=1
      else:
        dict_sens[i]=1
  return dict( sorted(dict_sens.items(), key=operator.itemgetter(1),reverse=True))
```

```python
sense_Nether=sensitive_eda(df_N)
sense_India=sensitive_eda(df_India)
sense_Japan=sensitive_eda(df_Japan)
sense_US=sensitive_eda(df)
print(sense_Nether)
print(sense_Japan)
print(sense_India)
print(sense_US)
```

```python
def plot_sense(col):
  countries=['India','Japan','USA','Netherlands']
  values=[sense_India[col],sense_Japan[col],sense_US[col],sense_Nether[col]]
  fig,ax=plt.subplots(figsize=(6,4))
  plt.bar(countries,values)
  ax.set_xlabel('Number of '+col)
  ax.set_ylabel('Frequency')
  ax.set_title(col+ ' Bar Graph')
  plt.legend(loc="upper right")
  # plt.xlim([0,])
  plt.show()
```
